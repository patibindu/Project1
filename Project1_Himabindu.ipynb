```python
import pandas as pd
from pathlib import Path
import numpy as np
import os
import matplotlib.pyplot as plt
from scipy.stats import linregress
import requests
import time
```

```python
monthly_factors = Path("Resources/Monthly_Macroeconomic_Factors.csv")
monthly_factors_pd = pd.read_csv(monthly_factors)
monthly_factors_pd = monthly_factors_pd.set_index("Date")
monthly_factors_pd.head()
```

```python
house_price_index = monthly_factors_pd["House_Price_Index"]
line_graph_House_Price = house_price_index.plot(kind = "line",figsize = (10,10))

line_graph_House_Price.set_title("House Price Index by Month")
line_graph_House_Price.set_ylabel("House Price Index")
line_graph_House_Price.set_xlabel("Date")

line_graph_House_Price
```

```python
Mortgage_Rate = monthly_factors_pd["Mortgage_Rate"]
line_graph_mortgage_rate = Mortgage_Rate.plot(kind = "line",figsize = (10,10))

line_graph_mortgage_rate.set_title("Mortgage Rate by Month")
line_graph_mortgage_rate.set_ylabel("Mortgage Rate")
line_graph_mortgage_rate.set_xlabel("Date")

line_graph_mortgage_rate
```
